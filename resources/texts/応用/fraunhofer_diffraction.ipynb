```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
l = 200               # スクリーンまでの距離 (mm)
wavelength = 5e-4     # 波長 λ (mm)
b = 1e-2              # スリット幅 b (mm)
q = 2 * np.pi / wavelength  # 波数 q
```

```python
y = np.linspace(-20, 20, 2000)  # mm
```

```python
phase = q * b * y / (2 * l)
```

```python
I_y = (b**2) * (np.sin(phase) / phase)**2
I_y[np.isnan(I_y)] = b**2
```

```python
plt.figure(figsize=(10, 6))
plt.plot(y, I_y, label=r"$I(y)$", color='orange')
plt.xlabel("y (mm)", fontsize=12)
plt.ylabel("Intensity I(y)", fontsize=12)
plt.title("Fraunhofer Diffraction Intensity Pattern", fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.axhline(0, color='black', linewidth=0.5, alpha=0.7)
plt.legend(fontsize=12)
plt.show()
```
